```python
import numpy as np

x = [1, 5, 6, 2, 8, 3, 9, 4, 6]

print("statistics for x")

mean_x = np.mean(x)
print("\nMean: ", mean_x)

std_x = np.std(x)
print("\nstd: ", std_x)

var_x = np.var(x)
print("\nvariance: ", var_x)

n = len(x) 
x.sort() 

if n % 2 == 0: 
    median1 = x[n//2] 
    median2 = x[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = x[n//2] 
print("\nMedian is: ",median)
```
